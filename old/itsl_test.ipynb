```python
from Lambert import *

test = lambda g, acc, ss : all(scan(g, w) == acc(w) for w in ss)
```

```python
lacc = lambda w: '010' not in w and '>10' not in '>'+w and '01<' not in w+'<' and '0110' not in w and '>110' not in '>'+w and '011<' not in w+'<' and w!='1' and w != '11' #This grammar accepts words where 1s are not solo: they must have an adjacent 1
import re
lacc = lambda w : all((lones := len(ones)) == 0 or lones >=3 for ones in re.split('0+', w))

lss = star('01', 16) #sigma star up to length 6
lss = [w for w in lss]
lsam = [w for w in lss if lacc(w)] #sample of positive data up to length 6
```

```python
g21 = learn(lsam) #learn with default values of k=2 and m=1. This should fail because this grammar is itsl

print(test(g21, lacc, lss))
g21
```
